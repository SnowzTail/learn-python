# Conclusion

## Quotations

*Premature optimisation is the root of all evil* - Donald Knuth

*Rules of optimisation:*

1.   *Don't*
2.   *(For experts only!) Don't yet*

-Michael A. Jackson

*The best is the enemy of the good* - Voltaire

The takeaway message of all these quotations is that optimisation is a double-edged sword. It can help make great code, but overdoing it or doing it too early can be an act of self-sabotage as it can detract from other things we want to achieve when we set out to write a code.

## Strategy

*   Write code that works
*   Write tests to ensure it keeps working as you change it
*   Profile it
*   Optimise the important bits of it

Considerations:

*   Time spent programming
*   Readability of code
*   Extendability of code
*   Complexity
*   Specificity of function
*   Testability
*   Learning







## An Example

If you're interested in reading about a real-world example of profiling and optimisation in practice, a nice story emerged regarding Grand Theft Auto Online in early 2021. Despite being a central part of Grand Theft Auto V, the [most profitable piece of media ever](https://www.gamesindustry.biz/articles/2018-04-09-gta-v-is-the-most-profitable-entertainment-product-of-all-time), it has been plagued with extreme loading times for the best part of a decade. Eventually, one player and skilled developer took the time to profile and optimise the loading process and cut loading times by about 70%. This [informative write-up](https://nee.lv/2021/02/28/How-I-cut-GTA-Online-loading-times-by-70/) goes a bit beyond what we cover in this course but, hopefully, you should still be able to see some parallels with what we've discussed in this course. This shows that, even for the biggest companies in the world, profiling and optimisation is extremely important.
