### The tasks

The file `exam_results.xlsx` contains data about some (fictional) students taking modules in a (fictional) Physics department.

1. Load the data with pandas. What are the observations and variables in this table?

```python
data = pd.read_excel("../assets/exam_results.xlsx")
data
```

2. Tidy the data.

```python
name = data.columns[:-2]
name
```

```python
df = data.melt(id_vars=name, var_name="Year", value_name="Grade")
df
```

```python
df.dropna(subset="Grade", inplace=True)
df
```

```python
df["Year"] = df["Year"].astype(int)
df
```

```python
programme = df["Degree Programme"].str.split(' ', expand=True)
programme
```

```python
pg = programme.drop(columns=2)
pg.columns = ["Degree", "Major", "Minor"]
pg
```

```python
df_ = df.drop(columns="Degree Programme").join(pg)
df_
```

```python
cols = df_.columns.tolist()
cols
```

```python
cols_ = cols[:3] + cols[-3:] + cols[3:-3]
cols_
```

```python
df_ = df_[cols_]
df_
```

3. Using your tidied data, answer the following questions:

* For MSci students, what is the mean grade for the Relativity exam?

```python
target = df_.query('Degree == "MSci" and Module == "Relativity"')
```

```python
target["Grade"].mean()
```

* What proportion of the students are taking Physics as their major subject?

```python
students = df_.iloc[:, :6].drop_duplicates()
students
```

```python
students.query('Major == "Physics"').shape[0] / students.shape[0]
```
